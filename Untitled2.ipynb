```python
import datetime
import unittest

def my_brand(hw_assignment):
    name = "Karthik Darapaneni"
    course = "Course 2024F-SSW567-A"
    pattern = "=*=*=*=%s=*=*=*="
    now = datetime.datetime.now().strftime("%d/%m/%Y %H:%M:%S")

    print(pattern % name)
    print(pattern % course)
    print(pattern % hw_assignment)
    print(pattern % now)

my_brand('Hw 01')
print('')

def classify_triangle(a, b, c):
    # Ensure inputs are valid lengths for triangle sides
    if a <= 0 or b <= 0 or c <= 0 or (a + b <= c) or (a + c <= b) or (b + c <= a):
        return "Invalid"
    
    # Check for right triangle using Pythagorean theorem
    right_triangle = (
        round(a**2 + b**2, 5) == round(c**2, 5) or
        round(a**2 + c**2, 5) == round(b**2, 5) or
        round(b**2 + c**2, 5) == round(a**2, 5)
    )
    
    # Check for triangle types
    if a == b == c:
        return "Equilateral"
    elif a == b or b == c or a == c:
        return "Isosceles and Right" if right_triangle else "Isosceles"
    else:
        return "Scalene and Right" if right_triangle else "Scalene"

class TestTriangleClassification(unittest.TestCase):
    def test_triangle(self):
        test_cases = [
            (5, 5, 5, "Equilateral"),
            (3, 3, 4, "Isosceles"),  # Example for isosceles triangle
            (3, 3, 4.24, "Isosceles and Right"),  # Example for isosceles and right triangle
            (6, 8, 10, "Scalene and Right"),  # Example for scalene and right triangle
            (25, 25, 25, "Equilateral"),
            (1, 2, 3, "Invalid"),
            (-1, 2, 3, "Invalid"),
            (7, 9, 12, "Scalene")  # Example for scalene triangle
        ]
        
        for a, b, c, expected in test_cases:
            result = classify_triangle(a, b, c)
            print(f"Testing classify_triangle({a}, {b}, {c}) => Expected: {expected}, Got: {result}")
            self.assertEqual(result)

if __name__ == "__main__":
    unittest.main(argv=[''], exit=False)  # Use this for running in Jupyter or similar environments
```
